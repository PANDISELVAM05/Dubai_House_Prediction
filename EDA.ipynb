```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load dataset
df_cleaned = pd.read_csv(r"D:\Labmentix\Project1\Dubai_House_Prediction.csv")

# Select categorical (string) and numeric columns
cat_cols = df_cleaned.select_dtypes(include=['object']).columns
num_cols = df_cleaned.select_dtypes(include=['int64', 'float64']).columns

# Encode categorical columns using LabelEncoder
le = LabelEncoder()
encoded_cats = pd.DataFrame()

for col in cat_cols:
    encoded_cats[col] = le.fit_transform(df_cleaned[col].astype(str))

# Combine numeric + encoded categorical columns
combined_df = pd.concat([df_cleaned[num_cols], encoded_cats], axis=1)

# Compute correlation matrix
corr_matrix = combined_df.corr(numeric_only=True)

# Extract only numeric rows vs categorical columns
corr_subset = corr_matrix.loc[num_cols, cat_cols]

# Plot heatmap
plt.figure(figsize=(12, 8))
# sns.heatmap(corr_subset, annot=True, cmap="YlGnBu", fmt=".2f")
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
# sns.heatmap(corr_subset, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Numeric Columns and Categorical Columns")
plt.xlabel("Categorical Columns")
plt.ylabel("Numeric Columns")
plt.show()
```
